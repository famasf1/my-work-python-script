```python
import pandas as pd
import openpyxl as pyxl
import tkinter as tk
from tkinter.filedialog import askopenfilename
```

```python
root = tk.Tk()
root.excel = pd.read_excel(askopenfilename(title="เลือกไฟล์ Excel",filetypes=(('Excel Files' , '*.xls'),('All Files','*.*'))))
root.withdraw()
```

```python
data = root.excel
```

```python
CCN = pd.Series(data['CCN'])
CCN_Insure = data[data["CCN"].str.contains(r"(^PHYIDINSURE.*)-(33)",regex=True)]
remove_word = CCN_Insure['CCN'].replace(['PHYIDINSURE'],'', regex=True).str.split('-')
df_remove = pd.DataFrame(remove_word)
df_remove_as_table = df_remove[['ID','Branch','Box Num']] = pd.DataFrame(df_remove.CCN.to_list(), index=df_remove.index)
```

```python
frames = [CCN_Insure, df_remove_as_table]
complete_frame = pd.concat(frames, axis=1)
```
